```python
import os
import matplotlib.pyplot as plt
import joblib
import numpy as np
import pandas as pd

plt.style.use("fivethirtyeight")
```

```python
class Perceptron:
    def __init__(self, eta: float=None, epochs: int=None): #eta is the learning rate, epochs is the no. of iterations(default value is None)
        self.weights = np.random.randn(3) * 1e-4 #initialise the weights to be small as convergence is faster 
        self.eta = eta # learning rate
        self.epochs = epochs # iterations
        
    def _z_outcome(self, inputs, weights): #Private method (convention used in python)
        return np.dot(inputs, weights) #multiply inputs with the corresponding weights 
    
    def activation_function(self, z):
        return np.where(z > 0, 1, 0) #if z > 0 put as 1,otherwise 0
        
    def fit(self, X, y):#X independent variable and y is the dependent variable
        self.X = X
        self.y = y
        
        X_with_bias = np.c_[self.X, -np.ones((len(self.X), 1))] #X-matrix with bias. 
        #np.c_ concatenation of matrixes with the 4x1 matrix of -1s
        print(f"X with bias: \n{X_with_bias}")
        
        for epoch in range(self.epochs):
            print("--"*10)
            print(f"for epoch >> {epoch + 1}")
            print("--"*10) #We do early stopping of epochs if the errors are 0 using if-else condition(we don't want to use unnecessary computing resources)
            
            z = self._z_outcome(X_with_bias, self.weights)
            y_hat = self.activation_function(z)
            print(f"predicted value after forward pass: \n{y_hat}")
            
            self.error = self.y - y_hat
            print(f"error: \n{self.error}")
            
            self.weights = self.weights + self.eta * np.dot(X_with_bias.T, self.error)
            print(f"updated weights after epoch: {epoch + 1}/{self.epochs}: \n{self.weights}")
            print(f"##"*10)
            
            
    def predict(self, X):
        X_with_bias = np.c_[X, -np.ones((len(X), 1))] #define X with bias matrix again
        z = self._z_outcome(X_with_bias, self.weights) #self.weights are the updated weights
        return self.activation_function(z)
```

```python
OR = {       #OR gate
    "x1": [0,0,1,1],
    "x2": [0,1,0,1],
    "y" : [0,1,1,1]
}

df_OR = pd.DataFrame(OR)

df_OR
```

```python
def prepare_data(df, target_col="y"):
    X = df.drop(target_col, axis=1)
    y = df[target_col]
    
    return X, y
```

```python
X, y = prepare_data(df_OR)
X
```

```python
y
```

```python
X, y = prepare_data(df_OR)

ETA = 0.1
EPOCHS = 10

model_or = Perceptron(eta=ETA, epochs=EPOCHS)

model_or.fit(X, y) #weights have converged in earlier epochs (Could have used early stopping)
```

```python
model_or.predict(X=[[0,0]]) #prediction from the perceptron algorithm 
```
